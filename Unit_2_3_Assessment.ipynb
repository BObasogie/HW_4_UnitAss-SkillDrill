````python
bitcoin = pd.read_csv('../Unsolved/bitcoin_cash_price.csv')
dash = pd.read_csv('../Unsolved/dash_price.csv')
crypto =  pd.merge(bitcoin, dash, on="Date")
crypto.head()
````
